```python
def start_hangman():
    import random
    myfile = open('words.txt', 'r')
    content = myfile.read()
    words = content.split()

    def len_words():
        lengths = []
        for i in words:
            if len(i) not in lengths:
                lengths.append(len(i))
        return lengths

    def list_words():
        wordlist3 = []
        wordlist4 = []
        wordlist5 = []
        wordlist6 = []
        for i in words:
            if len(i) == len_words()[0]:
                wordlist3.append(i)
            if len(i) == len_words()[1]:
                wordlist4.append(i)
            if len(i) == len_words()[2]:
                wordlist5.append(i)
            if len(i) == len_words()[3]:
                wordlist6.append(i)
        return wordlist3, wordlist4, wordlist5, wordlist6

    def dict_words():
        d = {}
        d['3'] = list_words()[0]
        d['4'] = list_words()[1]
        d['5'] = list_words()[2]
        d['6'] = list_words()[3]
        return d
    print('Welcome to HANGMAN game!!!')
    lengthallowed = ['3', '4', '5', '6']
    userinput = input('Enter word length : ')
    while userinput not in lengthallowed:
        userinput = input("Enter the length of the word [3-6] : ")
    if userinput == '3':
        word = random.choice(dict_words()['3'])
        print('Word : ', '*'*3)
    elif userinput == '4':
        word = random.choice(dict_words()['4'])
        print('Word : ', '*'*4)
    elif userinput == '5':
        word = random.choice(dict_words()['5'])
        print('Word : ', '*'*5)
    elif userinput == '6':
        word = random.choice(dict_words()['6'])
        print('Word : ', '*'*6)
    attempts = 2*len(word)-2
    guesses = ''
    wrong = 0
    wronglist = []
    while attempts > 0:
        print('\nAttempts remaining : ', attempts)
        print('Previous wrong guesses : ', wronglist)
        userinput1 = (input('Please guess a letter for this word : ')).lower()

        def check():
            c = 0
            for i in userinput1:
                c += 1
            return(c)
        while userinput1.isalpha() is not True or check() > 1:
            if userinput1.isalpha() is not True:
                userinput1 = input('Entered guess should be in range of [A-Z] or [a-z]')
            if check() > 1:
                userinput1 = input('Entered guess should be a single letter : ')
        if userinput1 in guesses:
            print("\nYou already guessed this letter, please try another letter")
            continue
        guesses += userinput1
        if userinput1 not in word:
            wrong += 1
            attempts -= 1
            wronglist.append(userinput1)
            print('\nLetter', userinput1, 'is not in the word')
        if userinput1 in word:
            print('\nLetter', userinput1, 'is in the word')
        print('Word : ', end='')
        for char in word:
            if char in guesses:
                print (char, end='')
            else:
                print('*', end='')

        def match():
            result = ''
            for i in word:
                if i in guesses:
                    result += i
            return result
        if word == match():
            print('\n\nCongratulations.You have guessed the right word.\nYou are safe')
            break
        if attempts == 0:
            print('\n\nYou have run out of tries.The word is', word.upper(), '.', '\nYou are HANGED.')
while True:
    start_hangman()
    again = input("\nDo you want to play again,Y/N? ").lower()
    if again.startswith("n"):
        print('\nOk.\nGame exited!')
        break
    print("\nYaay again!")
```
